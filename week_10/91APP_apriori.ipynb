### 載入所用package 及 91APP訂單資料
用mlxtend包來進行apriori分析

```python
import pandas  as  pd
from mlxtend.frequent_patterns  import apriori
from mlxtend.frequent_patterns import association_rules
```

```python
main_data = pd.read_csv('Ntu_Orders.csv')
```

```python
main_data.info()
```

### 顧客購物類別關聯性
製作每個顧客ID曾經購買的商品類別表

```python
data_apriori= main_data.groupby(['MemberId', 'LevelFourCategoryName'])['Quantity'].sum().unstack(fill_value=0).astype('bool')
data_apriori.head()
```

進行分析 發現 T恤/帽T 與其他衣物的搭配經常出現

```python
apriori_result = apriori(data_apriori, min_support=0.07, use_colnames=True)
rules = association_rules(apriori_result, metric="lift", min_threshold=1)
rules=rules[rules["confidence"]>0.6]
rules
```

對較高層級Levelthree的商品類別作分析 不過並無特別的發現

```python
main_data['LevelThreeCategoryName'].unique()
```

```python
data_apriori =main_data.groupby(['MemberId', 'LevelThreeCategoryName'])['Quantity'].sum().unstack(fill_value=0).astype('bool')
apriori_result = apriori(data_apriori, min_support=0.07, use_colnames=True)
rules = association_rules(apriori_result, metric="lift", min_threshold=1)
rules=rules[rules["confidence"]>0.6]
rules
```

### 顧客對特定商品類別購買關聯性
貼身衣物的情況 其中舒適/日系這對組合較重要

```python
underwear_data = main_data.groupby([ 'LevelTwoCategoryName']).get_group('內衣')
```

```python
data_apriori = underwear_data.groupby(['MemberId', 'LevelFourCategoryName'])['Quantity'].sum().unstack(fill_value=0).astype('bool')
apriori_result = apriori(data_apriori, min_support=0.01, use_colnames=True)
rules = association_rules(apriori_result, metric="lift", min_threshold=1)
rules
```
