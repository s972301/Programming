### 2016總統就職演講稿 詞頻分析


```python
import jieba
import sys
import scipy
import sklearn
import pandas as pd
import io

from sklearn import feature_extraction
from sklearn.feature_extraction.text import TfidfVectorizer
```

### 讀取講稿txt
每句話分開來讀取

```python
f = io.open('presidentSpeech.txt', 'r',encoding = 'UTF-8')
text=list()

for line in io.open('presidentSpeech.txt',encoding = 'UTF-8'):  
    line = f.readline()    
    text.append(" ".join(jieba.cut(line.split(',')[0], cut_all=False)))  
```

### 詞頻分析
以每一句話為單位計算Tf_Idf

```python
vectorizer = TfidfVectorizer()
tfidf = vectorizer.fit_transform(text)
print tfidf.shape

words = vectorizer.get_feature_names()
for i in xrange(len(text)):
    print '----Document %d----' % (i)
    for j in xrange(len(words)):
        if tfidf[i,j] > 1e-5:
              print words[j].encode('utf-8'), tfidf[i,j]
```
